```python
from tkinter import *
import random

# list of colours
colours = ['Red','Blue','Green','Pink','Black',
           'White','Orange','Purple']


score = 0

# timer 
secondsleft = 40

def startGame(event):

    if secondsleft == 40:
       
        countdown()
    
    nextColour()

    
def nextColour():
   
    global score
    global secondsleft
    
    if secondsleft> 0:

        entry.focus_set()

        if entry.get().lower() == colours[1].lower():

            score=score+1
        
        entry.delete(0,END)

        random.shuffle(colours)

        label.config(fg = str(colours[1]), text = str(colours[0]))
        
        scoreLabel.config(text = "Score: " + str(score))



def countdown():

    global secondsleft

    if secondsleft> 0:

        secondsleft=secondsleft-1
        timeLabel.config(text = "Seconds left: "+ str(secondsleft))
        timeLabel.after(1000, countdown)


# GUI window
root = Tk()


root.title("TYPE SPEED TEST")
root.geometry("900x600")
root.configure(background='Yellow')


instructions = Label(root, text = "Type in the colour "
                        "of the words and not the word text!",font = ('Times', 28),background="White", foreground="Black")
instructions.pack() 

scoreLabel = Label(root, text = "Press ENTER KEY to start",font = ('Times', 24),background="Orange", foreground="Brown")
scoreLabel.pack()

luckLabel = Label(root, text = "GOOD LUCK!",font = ('Times', 24),background="Orange", foreground="Purple")
luckLabel.pack()

timeLabel = Label(root, text = "Seconds left: " +str(secondsleft), font = ('Times', 24),background="Orange", foreground="Brown")

timeLabel.pack()

label = Label(root, font = ('Times', 100))
label.pack()

entry = Entry(root)

root.bind('<Return>', startGame)
entry.pack()

entry.focus_set()

#quit command 
Button(root,text="quit",command=root.destroy).pack()

root.mainloop()
```
